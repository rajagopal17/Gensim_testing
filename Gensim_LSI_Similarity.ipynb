```python
doc_file2=nlp(open('arnav.txt').read())
sentence=list(doc_file2.sents)
#sentence
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence),deacc=True,min_len=2))
        
data_input=list(sent_to_words(sentence))
#data_input
```

```python
#Convert entire text into tokens
#Assign a unique number to each token/word using Dictionary function
#Dictionary of words need to be converted to bag of words -doc2bow-(id of the word from dictionary , number of times the word appears in the corpus)
#doc2bow need to converted into a corpus (Market matrix format)
```

```python
from gensim.models import TfidfModel
from gensim.corpora import Dictionary

#dataset = api.load("text8")
dct = Dictionary(data_input)  # fit dictionary
mycorpus = [dct.doc2bow(line) for line in data_input]  # convert corpus to BoW format

mymodel = TfidfModel(mycorpus)  # fit model
vector = mymodel[mycorpus]# apply model to the first corpus document
vector
```

```python
print(dct.token2id)#list of all tokens with their unique id
#or 
dct.token2id.values() #same as above
dct.token2id.keys()#only unique ids
dct.token2id.items()#only the tokens
```

```python
lsi = models.LsiModel(mycorpus, id2word=dct, num_topics=300)  # initialize an LSI transformation
my_lsi = lsi[vector]  # create a double wrapper over the original corpus: bow->tfidf->fold-in-lsi
```

```python
#To test a new string of text with the existing dictionary
test = "khushi was not"
test_trf = dct.doc2bow(test.lower().split())
#print(new_vec) 
```

```python
test_lsi=lsi[test_trf]
```

```python
index = similarities.MatrixSimilarity(lsi[mycorpus]) 
```

```python
simi = index[test_lsi]  # perform a similarity query against the corpus
```

```python
simi = sorted(enumerate(simi),key=lambda item: -item[1])
```

```python
similarity=list(simi)
pd.DataFrame(similarity[0:10])
```
